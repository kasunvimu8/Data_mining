```python
#random walker

from numpy.random import randint

class RandomWalk:
    def __init__(self, max_steps):
        self.max_steps = max_steps

    def walk(self):
        random_numbers = randint(0, 2, size=self.max_steps)
        steps = np.where(random_numbers == 0, -1, +1)
        path = np.cumsum(steps)
        return path

random_walker = RandomWalk(500)
Pos_v1 = random_walker.walk()
Pos_v2 = random_walker.walk()

fig = plt.figure(figsize=(18,4))
fig.subplots_adjust(hspace=0.05)
axes1= fig.add_subplot(131)
axes1.plot(Pos_v1)
axes1.set_title("Normal plot")

axes2= fig.add_subplot(132)
axes2.hist(Pos_v1 , bins=20, color='r')
axes2.set_title("histogram plot")

axes3= fig.add_subplot(133)
axes3.scatter(Pos_v1,Pos_v2 , color='green', alpha=0.3, edgecolors='grey')
axes3.set_title("scatter plot")

line_labels = ["normal", "histogram", "scatter"]


fig.legend( labels=line_labels,   # The labels for each line
           loc="center right",   # Position of legend
           borderaxespad=0.1,    # Small spacing around legend box
           title="All subplots legend "  # Title for the legend
           )


plt.show()
```
